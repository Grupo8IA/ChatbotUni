# Obtención de data: Agrupación de datos, conteo de palabras y caracteres, eliminación de duplicados.

```python
import re
import pandas as pd
```

Conteo de palabras y caracteres válidos usando `re`.

```python
def limpiar(file, lista, maximo):
    f = open(file+".txt", encoding='utf-8')
    for linea in f:
        rsp = linea.split('\t')
        regex = "(([\wáéíóúñ]+)|[\.\,\"\'\?\¿\!\¡])(\\t|\\n)?"

        if len(rsp) == 2 :
            nropregunta = len(re.findall(regex, rsp[0]))
            nrorespuesta = len(re.findall(regex, rsp[1]))
            if nropregunta <= maximo and nropregunta <= maximo :
                lista.append(linea)
    f.close()
    return lista
```

Eliminación de Duplicados

```python
def duplicados(lista):
    df = pd.DataFrame(lista, columns=['linea'])
    dprow = df.drop_duplicates()
    final = dprow.values.tolist()
    print(len(final))
    return lista
```
